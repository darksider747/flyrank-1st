```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.

import pandas as pd
df = pd.read_csv("data/raw/content_refresh_anonymized.csv")
print(f"Rows: {len(df)}, Columns: {df.shape[1]}")
print(f"Unique clients: {df['client_id'].nunique()}")
```

## 2. The question: decision, action, cost of a wrong call

Unit of analysis: one web page (one row = one page, identified by content_id).

Decision this improves: which pages a content team should review first,
since they can't check every page.

Action taken: a reviewer opens the top pages on my ranked list and decides
whether to rewrite, expand, or update them.

Cost of a wrong call: flagging a fine page wastes limited reviewer time.
Missing a truly declining page means it keeps losing traffic silently.

Why data/ML helps: a plain rule already found some declining pages, but a
trained model found roughly 3x more of them correctly in the same top-50
spots — real evidence a smarter approach adds value over guessing.

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
declining_count = (df["trend_direction"] == "down").sum()
print(f"Pages currently marked as declining: {declining_count} out of {len(df)}")
```

## 3. Quick look at the data (2-3 real numbers)

High Decline Rate: Over half of the web pages (54%) in this dataset are losing traffic. Because this issue affects so many pages, a human content team cannot possibly check them all. They genuinely need a smart checklist to prioritize their work.Massive CTR Differences: Your positions matter heavily. Pages ranking high on the first page have an average click-through rate (CTR) of 0.652, while deeper rankings drop to just 0.150. This means top rankings get over 4 times more clicks, proving that even small shifts in rank have huge traffic consequences.Data Quality Reality Check: The data has a weird quirk where the top_3 tier shows an impossible CTR of over 100% across 2,321 pages. Instead of ignoring this or letting it break the model, you safely bypassed it by relying on cleaner, realistic data groups (page_1 and deep). This shows there is a strong, real signal the model can learn from.


```python
declining_share = (df["trend_direction"] == "down").mean()
ctr_page1 = df[df["position_tier"] == "page_1"]["ctr"].mean()
ctr_deep = df[df["position_tier"] == "deep"]["ctr"].mean()

print(f"Share of pages declining: {declining_share:.1%}")
print(f"Mean CTR at page_1 position: {ctr_page1:.3f}")
print(f"Mean CTR at deep position: {ctr_deep:.3f}")
```
